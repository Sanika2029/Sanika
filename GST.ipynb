```python
X.fillna(method='ffill', inplace=True)
```

```python
X.nunique()
```

```python
numerical_cols = X.columns[[0, 1, 2, 3, 4, 5, 6, 7, 8, 13, 15, 16, 14]]
categorical_cols = X.columns[[9, 10, 11, 12, 19, 18, 17]]
```

```python



from sklearn.preprocessing import StandardScaler, OneHotEncoder
scaler = StandardScaler()

# Scale the numerical columns
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse=False)  # set sparse=False to get a dense array

# One-hot encode the categorical columns
encoded_cols = encoder.fit_transform(X[categorical_cols])

# Convert the encoded array to a DataFrame
encoded_df = pd.DataFrame(encoded_cols, columns=[f'Column{col}_encoded' for col in range(encoded_cols.shape[1])])

# Concatenate the scaled numerical columns and one-hot encoded categorical columns
X_preprocessed = pd.concat([X[numerical_cols], encoded_df], axis=1)

# Print the preprocessed dataset
print(X_preprocessed.head(5))
```

```python
X_preprocessed
```

```python
from sklearn.decomposition import PCA

# Create a PCA object with 5 components (you can adjust this number)
pca = PCA(n_components=5)

# Fit and transform the data
X_pca = pca.fit_transform(X_preprocessed)

print(X_pca.shape)  # Output: (785133, 5)
```

```python
y_new = y.drop('ID', axis=1)
y_new
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_pca, y_new, test_size=0.2, random_state=42)

# Create a Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
rfc.fit(X_train, y_train)

# Make predictions on the validation data
y_pred_val = rfc.predict(X_val)

# Evaluate the model on the validation data
print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))
print("Validation Classification Report:")
print(classification_report(y_val, y_pred_val))
print("Validation Confusion Matrix:")
print(confusion_matrix(y_val, y_pred_val))
```

```python
# Import necessary libraries
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_pca, y_new, test_size=0.2, random_state=0)

# Create a Random Forest Classifier
rfc = MultinomialNB(alpha=0.1)

# Train the model on the training data
rfc.fit(X_train, y_train)

# Make predictions on the validation data
y_pred_val = rfc.predict(X_val)

# Evaluate the model on the validation data
print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))
print("Validation Classification Report:")
print(classification_report(y_val, y_pred_val))
print("Validation Confusion Matrix:")
print(confusion_matrix(y_val, y_pred_val))
```
